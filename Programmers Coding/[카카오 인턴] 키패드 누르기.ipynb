```python
def solution(numbers,hand):
    
    key =  [[1,2,3],
           [4,5,6],
           [7,8,9],
           ["*",0,"#"]]
    
    result = ""
    
    start_left = (3,0)
    start_right = (3,2)
    
    left_pad = [1,4,7]
    right_pad = [3,6,9]
    center_pad = [2,5,8,0]
    
    for i in numbers:
        
        for j in range(4):
            
            for k in range(3):
                
                if i == key[j][k]:
                    
                    n_row = j
                    n_col = k
                    
                    
                    
                    if i in left_pad:
                        
                        result += "L"
                        start_left = (j,k)
                    
                    if i in right_pad:
                        
                        result += "R"
                        start_right = (j,k)
                        
                    if i in center_pad:
                        
                        _left = (abs(start_left[0]-j),abs(start_left[1]-k))
                        _right = (abs(start_right[0]-j),abs(start_right[1]-k))
                        _left_sum = _left[0] + _left[1]
                        _right_sum = _right[0] + _right[1]
                        
                        
                        if _left_sum < _right_sum:
                        
                            result += "L"
                            start_left = (j,k)
                        
                        elif _left_sum > _right_sum:
                        
                            result += "R"
                            start_right = (j,k)
                            
                        else:
                        
                            if hand == "left":
                            
                                result += "L"
                                start_left = (j,k)
                        
                            if hand == "right":
                            
                                result += "R"
                                start_right = (j,k)
                    
       
                        

    return result
                
                    
                    
                    
                    
                    
```
